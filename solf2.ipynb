## Descripción General del Dataset de Ventas de Productos Electrónicos

Este dataset, denominado `ventas_electronica.csv`, es un conjunto de datos simulado que representa transacciones de venta de una tienda de electrónica. Fue creado específicamente para fines educativos, para proporcionar un escenario realista que incluya datos faltantes y duplicados, permitiendo así una práctica integral en limpieza, manipulación y análisis de datos. No proviene de una fuente real o pública, sino que ha sido confeccionado para replicar las características comunes de un dataset de ventas.

## Estructura y Significado de las Columnas

El dataset contiene información detallada sobre cada venta realizada. A continuación, se describe cada columna:

- `ID_Venta`: Identificador único de cada transacción de venta. Es un número entero. Tipo de Datos Esperado: Numérico (entero).

- `Fecha`: La fecha en que se realizó la venta. Tipo de Datos Esperado: Fecha/Hora.

- `Producto`: Nombre del producto vendido (e.g., "Laptop Ultra", "Auriculares Bluetooth"). Tipo de Datos Esperado: Categórico/Texto (string).

- `Categoria`: La categoría a la que pertenece el producto (e.g., "Portátiles", "Audio", "Smartphones"). Tipo de Datos Esperado: Categórico/Texto (string).

- `Cantidad`: Número de unidades del producto vendidas en esa transacción.Tipo de Datos Esperado: Numérico (entero).

- `Precio_Unitario`: El precio de venta original por una sola unidad del producto, antes de aplicar cualquier descuento. Tipo de Datos Esperado: Numérico (flotante).

- `Costo_Unitario`: El costo incurrido por la tienda para adquirir una sola unidad de ese producto. Tipo de Datos Esperado: Numérico (flotante).

- `Metodo_Pago`: El método utilizado por el cliente para pagar la compra (e.g., "Tarjeta", "Efectivo", "Transferencia"). Tipo de Datos Esperado: Categórico/Texto (string).

- `Ciudad_Venta`: La ciudad donde se realizó la venta (e.g., "Madrid", "Barcelona"). Tipo de Datos Esperado: Categórico/Texto (string).

- `Cliente_Edad`: La edad del cliente que realizó la compra. Tipo de Datos Esperado: Numérico (entero).

- `Cliente_Genero`: El género del cliente ( "M" para masculino, "F" para femenino). Tipo de Datos Esperado: Categórico/Texto (string).

- `Descuento_Aplicado`: El porcentaje de descuento aplicado a esta venta, expresado como una fracción (e.g., 0.05 para un 5%). Tipo de Datos Esperado: Numérico (flotante).

## Objetivo General del Análisis

El objetivo principal de analizar este dataset es extraer información valiosa sobre el rendimiento de ventas de la tienda de electrónica, identificar patrones y generar insights que puedan fundamentar decisiones de negocio. Esto incluye:

- Análisis Descriptivo: Comprender las características generales de las ventas, como la distribución de productos vendidos, los métodos de pago más comunes, y la demografía de los clientes.

- Identificación de Tendencias y Patrones: Descubrir patrones de venta a lo largo del tiempo (por ejemplo, tendencias mensuales), identificar las categorías de productos más vendidas o rentables, y entender el comportamiento de compra según la ubicación o las características del cliente.

- Generación de Insights Accionables: Utilizar los resultados del análisis para sugerir mejoras en estrategias de marketing, gestión de inventario, optimización de precios o expansión a nuevas ciudades, maximizando así la rentabilidad y la satisfacción del cliente.

En resumen, el análisis busca transformar datos brutos de ventas en conocimiento útil para mejorar el desempeño del negocio.

```python
import pandas as pd
import numpy as np # Necesario para np.nan, que representa valores nulos.
import matplotlib.pyplot as plt

# --- 1. Carga y Primera Vista: ---
# Cargar el dataset desde el archivo CSV.
# Pandas es inteligente y por defecto interpreta las celdas vacías como NaN (Not a Number), que es el valor de NumPy para datos faltantes.
try:
    df_ventas = pd.read_csv('ventas_electronica.csv')
    print("1. DataFrame cargado exitosamente. Primeras 7 filas:")
    print(df_ventas.head(7)) # Muestra las primeras 7 filas del DataFrame.
    print("-" * 50)
except FileNotFoundError:
    print("Error: El archivo 'ventas_electronica.csv' no se encontró.")
    print("Asegúrate de haber creado el archivo con el contenido proporcionado y en la misma ubicación que tu script Python.")
    # Crea un DataFrame de prueba si el archivo no se encuentra, para que los estudiantes no se detengan.
    print("\nCreando un DataFrame de prueba para continuar el ejercicio (¡ADVERTENCIA: este DF no simula la complejidad real de nulos/duplicados del archivo original!):")
    data_test = {
        'ID_Venta': range(1, 101),
        'Fecha': pd.to_datetime(pd.date_range(start='2023-01-01', periods=100).strftime('%Y-%m-%d')),
        'Producto': ['Laptop', 'Smartphone', 'Audífonos', 'Monitor', 'Teclado'] * 20,
        'Categoria': ['Portátiles', 'Smartphones', 'Audio', 'Monitores', 'Accesorios'] * 20,
        'Cantidad': [1, 2, 1, 3, 1] * 20,
        'Precio_Unitario': [1200.0, 850.0, 75.0, 300.0, 90.0] * 20,
        'Costo_Unitario': [800.0, 600.0, 40.0, 180.0, 50.0] * 20,
        'Metodo_Pago': ['Tarjeta', 'Efectivo', 'Transferencia', 'Tarjeta', 'Efectivo'] * 20,
        'Ciudad_Venta': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia'] * 25,
        'Cliente_Edad': list(range(20, 70)) * 2,
        'Cliente_Genero': ['M', 'F'] * 50,
        'Descuento_Aplicado': [0.05, 0.0, 0.10, 0.0, 0.05] * 20
    }
    df_ventas = pd.DataFrame(data_test)
    # Introducir algunos NaN y duplicados manualmente en el DF de prueba para simular el real:
    df_ventas.loc[5:10, 'Precio_Unitario'] = np.nan
    df_ventas.loc[15:20, 'Cliente_Edad'] = np.nan
    df_ventas.loc[25:30, 'Descuento_Aplicado'] = np.nan
    df_ventas = pd.concat([df_ventas, df_ventas.iloc[0:5]]).reset_index(drop=True)
    print("DataFrame de prueba creado con algunos nulos y duplicados. Continúa con los ejercicios.")
    print("-" * 50)

# --- 2. Dimensiones: ---
# Muestra el número de filas y columnas del DataFrame.
print("2. Dimensiones del DataFrame (Filas, Columnas):")
print(df_ventas.shape)
print("-" * 50)

# --- 3. Tipos de Datos: ---
# `info()` es excelente para obtener un resumen rápido de las columnas, sus tipos de datos
# y la cantidad de valores no nulos.
print("3. Tipos de Datos y Valores No Nulos:")
df_ventas.info()
print("-" * 50)

# --- 4. Resumen Descriptivo: ---
# `describe()` calcula estadísticas descriptivas como media, desviación estándar, min, max, etc.,
# solo para columnas numéricas por defecto.
print("4. Resumen Estadístico Descriptivo de Columnas Numéricas:")
print(df_ventas.describe())
print("-" * 50)

# --- 5. Valores Nulos: Conteo: ---
# `isnull()` crea un DataFrame booleano (True donde hay nulos, False donde no).
# `sum()` suma los True (que se tratan como 1) por columna, dando el conteo de nulos.
print("5. Cantidad de Valores Nulos en cada columna:")
print(df_ventas.isnull().sum())
print("-" * 50)

# --- 6. Valores Nulos: Porcentaje: ---
# Dividimos el conteo de nulos por el número total de filas y multiplicamos por 100.
print("6. Porcentaje de Valores Nulos en cada columna:")
print((df_ventas.isnull().sum() / len(df_ventas)) * 100)
print("-" * 50)

# --- 7. Manejo de Nulos (Cliente_Edad): ---
# Rellenamos los NaN en 'Cliente_Edad' con la mediana.
# La mediana es preferible a la media para edades, ya que es menos sensible a valores atípicos (outliers)
# y representa mejor el valor central en distribuciones asimétricas (que son comunes para edades).
print("7. Rellenando valores nulos en 'Cliente_Edad' con la mediana...")
median_edad = df_ventas['Cliente_Edad'].median() # Calculamos la mediana.
df_ventas['Cliente_Edad'].fillna(median_edad, inplace=True) # Usamos inplace=True para modificar el DataFrame directamente.
print(f"Mediana de 'Cliente_Edad' usada para rellenar nulos: {median_edad}")
print("-" * 50)

# --- 8. Manejo de Nulos (Precio/Costo/Descuento): ---
# Rellenamos los NaN en estas columnas con 0.
# Justificación: Para ventas, un precio, costo o descuento nulo probablemente significa que no hubo tal valor,
# o fue un error de entrada que podemos interpretar como 0 para cálculos financieros.
# Por ejemplo, un descuento nulo es equivalente a un descuento del 0%.
print("8. Rellenando valores nulos en 'Precio_Unitario', 'Costo_Unitario' y 'Descuento_Aplicado' con 0...")
df_ventas['Precio_Unitario'].fillna(0, inplace=True)
df_ventas['Costo_Unitario'].fillna(0, inplace=True)
df_ventas['Descuento_Aplicado'].fillna(0, inplace=True)
print("Nulos rellenados en las columnas especificadas.")
print("-" * 50)

# --- 9. Verificación de Nulos: ---
# Volvemos a contar los nulos para confirmar que las columnas objetivo ya no los tienen.
print("9. Verificación final de valores nulos (deberían ser 0 para las columnas procesadas):")
print(df_ventas[['Cliente_Edad', 'Precio_Unitario', 'Costo_Unitario', 'Descuento_Aplicado']].isnull().sum())
print("-" * 50)

# --- 10. Conversión de Tipo (Fecha): ---
# Convertimos la columna 'Fecha' al tipo datetime de Pandas.
# Esto es crucial para realizar operaciones basadas en tiempo (filtrar por mes, año, etc.).
print("10. Convirtiendo la columna 'Fecha' a tipo datetime...")
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])
print("Nuevo tipo de datos para 'Fecha':")
print(df_ventas['Fecha'].dtype)
print("-" * 50)

# --- 11. Detección de Duplicados: ---
# `duplicated()` devuelve una serie booleana indicando si cada fila es un duplicado de una fila anterior.
# `sum()` cuenta los True.
print("11. Número de filas completamente duplicadas encontradas:")
print(df_ventas.duplicated().sum())
print("-" * 50)

# --- 12. Eliminación de Duplicados: ---
# `drop_duplicates()` elimina las filas duplicadas.
# `inplace=True` modifica el DataFrame directamente.
# `keep='first'` asegura que se conserve la primera ocurrencia del duplicado.
print("12. Eliminando filas duplicadas (conservando la primera ocurrencia)...")
df_ventas.drop_duplicates(inplace=True)
print("Nueva cantidad de filas después de eliminar duplicados:")
print(df_ventas.shape[0]) # df.shape[0] da el número de filas.
print("-" * 50)
```

```python
# Parte II
# Asegúrate de que df_ventas esté cargado y limpio de la Parte 1.
# Si estás ejecutando este script en partes, puedes recargar el DataFrame limpio aquí, o simplemente continuar.

# --- 13. Cálculo de 'Precio_Venta_Neto': ---
# El precio neto es el precio original menos el descuento aplicado.
# Se usa (1 - Descuento_Aplicado) para directamente obtener el factor de precio después del descuento.
print("13. Calculando 'Precio_Venta_Neto'...")
df_ventas['Precio_Venta_Neto'] = df_ventas['Precio_Unitario'] * (1 - df_ventas['Descuento_Aplicado'])
print("Columna 'Precio_Venta_Neto' agregada. Primeras filas con la nueva columna:")
print(df_ventas[['Precio_Unitario', 'Descuento_Aplicado', 'Precio_Venta_Neto']].head())
print("-" * 50)

# --- 14. Cálculo de 'Ingreso_Total_Venta': ---
# El ingreso total es la cantidad de productos vendidos por su precio de venta neto.
print("14. Calculando 'Ingreso_Total_Venta'...")
df_ventas['Ingreso_Total_Venta'] = df_ventas['Cantidad'] * df_ventas['Precio_Venta_Neto']
print("Columna 'Ingreso_Total_Venta' agregada. Primeras filas con la nueva columna:")
print(df_ventas[['Cantidad', 'Precio_Venta_Neto', 'Ingreso_Total_Venta']].head())
print("-" * 50)

# --- 15. Cálculo de 'Costo_Total_Venta': ---
# El costo total es la cantidad de productos vendidos por su costo unitario.
print("15. Calculando 'Costo_Total_Venta'...")
df_ventas['Costo_Total_Venta'] = df_ventas['Cantidad'] * df_ventas['Costo_Unitario']
print("Columna 'Costo_Total_Venta' agregada. Primeras filas con la nueva columna:")
print(df_ventas[['Cantidad', 'Costo_Unitario', 'Costo_Total_Venta']].head())
print("-" * 50)

# --- 16. Cálculo de 'Ganancia_Bruta': ---
# La ganancia bruta es la diferencia entre el ingreso total y el costo total de la venta.
print("16. Calculando 'Ganancia_Bruta'...")
df_ventas['Ganancia_Bruta'] = df_ventas['Ingreso_Total_Venta'] - df_ventas['Costo_Total_Venta']
print("Columna 'Ganancia_Bruta' agregada. Primeras filas con la nueva columna:")
print(df_ventas[['Ingreso_Total_Venta', 'Costo_Total_Venta', 'Ganancia_Bruta']].head())
print("-" * 50)

# --- 17. Cálculo de 'Margen_Ganancia_Porcentaje': ---
# El margen de ganancia es la ganancia bruta dividida por el ingreso total, multiplicada por 100.
# Importante: Usamos `np.where` para evitar divisiones por cero.
# Si 'Ingreso_Total_Venta' es 0, el margen también es 0 para evitar errores.
print("17. Calculando 'Margen_Ganancia_Porcentaje'...")
df_ventas['Margen_Ganancia_Porcentaje'] = np.where(
    df_ventas['Ingreso_Total_Venta'] != 0, # Condición: si el ingreso total no es cero
    (df_ventas['Ganancia_Bruta'] / df_ventas['Ingreso_Total_Venta']) * 100, # Valor si la condición es True
    0 # Valor si la condición es False (es decir, si Ingreso_Total_Venta es 0)
)
print("Columna 'Margen_Ganancia_Porcentaje' agregada. Primeras filas con la nueva columna:")
print(df_ventas[['Ingreso_Total_Venta', 'Ganancia_Bruta', 'Margen_Ganancia_Porcentaje']].head())
print("-" * 50)

# --- 18. Filtrado por Categoría: ---
# Seleccionamos filas donde la columna 'Categoria' es exactamente 'Portátiles'.
print("18. Ventas solo de la categoría 'Portátiles':")
ventas_portatiles = df_ventas[df_ventas['Categoria'] == 'Portátiles']
print(ventas_portatiles.head())
print(f"Número de ventas de portátiles: {len(ventas_portatiles)}")
print("-" * 50)

# --- 19. Filtrado por Rango de Fechas: ---
# Filtramos usando el rango de fechas para el segundo trimestre de 2023 (Abril a Junio).
# Es importante que la columna 'Fecha' sea de tipo datetime para que esto funcione correctamente.
print("19. Ventas realizadas en el segundo trimestre de 2023 (Abril a Junio):")
ventas_segundo_trimestre = df_ventas[
    (df_ventas['Fecha'] >= '2023-04-01') & # Fecha es igual o posterior al 1 de abril de 2023
    (df_ventas['Fecha'] <= '2023-06-30')   # Fecha es igual o anterior al 30 de junio de 2023
]
print(ventas_segundo_trimestre.head())
print(f"Número de ventas en el segundo trimestre: {len(ventas_segundo_trimestre)}")
print("-" * 50)

# --- 20. Filtrado por Múltiples Condiciones: ---
# Combinamos múltiples condiciones usando el operador lógico AND (`&`).
print("20. Ventas de 'Audio' con Cantidad > 1 y Método de Pago 'Efectivo':")
filtro_complejo = df_ventas[
    (df_ventas['Categoria'] == 'Audio') &
    (df_ventas['Cantidad'] > 1) &
    (df_ventas['Metodo_Pago'] == 'Efectivo')
]
print(filtro_complejo.head())
print(f"Número de ventas que cumplen todas las condiciones: {len(filtro_complejo)}")
print("-" * 50)

# --- 21. Selección con .loc (Filas y Columnas): ---
# `.loc` permite seleccionar filas y columnas por sus etiquetas (nombres).
# Primero, la condición para las filas, luego una lista de nombres de columnas.
print("21. Ventas en 'Barcelona' con columnas específicas usando .loc:")
ventas_barcelona_info = df_ventas.loc[
    df_ventas['Ciudad_Venta'] == 'Barcelona', # Condición para seleccionar filas
    ['Fecha', 'Producto', 'Ingreso_Total_Venta', 'Ganancia_Bruta'] # Lista de columnas a seleccionar
]
print(ventas_barcelona_info.head())
print(f"Número de ventas en Barcelona: {len(ventas_barcelona_info)}")
print("-" * 50)

# --- 22. Selección con .iloc (Rango de Filas y Columnas): ---
# `.iloc` permite seleccionar filas y columnas por sus posiciones (índices numéricos).
# [inicio_fila:fin_fila, inicio_columna:fin_columna]
# Recuerda que el 'fin' es exclusivo.
print("22. Selección de filas (50 a 99) y columnas (Producto, Cantidad, Precio_Unitario) usando .iloc:")
# Para 'Producto', 'Cantidad', 'Precio_Unitario', necesitamos encontrar sus índices numéricos.
# Una forma pedagógica es saber que están secuencialmente en el dataset original o usar:
# df_ventas.columns.get_loc('Producto'), df_ventas.columns.get_loc('Precio_Unitario')
# Suponiendo el orden del CSV: Producto (2), Cantidad (3), Precio_Unitario (4)
# Así que los índices de columna serían del 2 al 5 (exclusivo).
seleccion_iloc = df_ventas.iloc[50:100, [df_ventas.columns.get_loc('Producto'),
                                         df_ventas.columns.get_loc('Cantidad'),
                                         df_ventas.columns.get_loc('Precio_Unitario')]]
print(seleccion_iloc.head())
print(f"Número de filas seleccionadas con .iloc: {len(seleccion_iloc)}")
print("-" * 50)
```

```python
# Asegúrate de que df_ventas esté cargado y con las columnas calculadas de las Partes 1 y 2.

# --- 23. Ventas Totales por Ciudad: ---
# Agrupamos por 'Ciudad_Venta' y luego sumamos 'Ingreso_Total_Venta' y 'Ganancia_Bruta'.
print("23. Ingresos y Ganancias Totales por Ciudad de Venta:")
ventas_por_ciudad = df_ventas.groupby('Ciudad_Venta')[['Ingreso_Total_Venta', 'Ganancia_Bruta']].sum()
print(ventas_por_ciudad)
print("-" * 50)

# --- 24. Ganancia Promedio por Categoría y Género: ---
# Agrupamos por dos columnas y calculamos la media de 'Ganancia_Bruta'.
print("24. Ganancia Bruta Promedio por Categoría de Producto y Género del Cliente:")
ganancia_por_cat_gen = df_ventas.groupby(['Categoria', 'Cliente_Genero'])['Ganancia_Bruta'].mean()
print(ganancia_por_cat_gen)
print("-" * 50)

# --- 25. Producto Top por Ingreso: ---
# Agrupamos por 'Producto', sumamos 'Ingreso_Total_Venta' y luego usamos `idxmax()` para encontrar el índice (producto) con el valor máximo.
print("25. Producto que ha generado el mayor Ingreso Total de Venta:")
producto_top_ingreso = df_ventas.groupby('Producto')['Ingreso_Total_Venta'].sum().idxmax()
ingreso_maximo = df_ventas.groupby('Producto')['Ingreso_Total_Venta'].sum().max()
print(f"El producto con el mayor ingreso total es: '{producto_top_ingreso}' con un ingreso de ${ingreso_maximo:,.2f}")
print("-" * 50)

# --- 26. Método de Pago Más Rentable: ---
# Agrupamos por 'Metodo_Pago', sumamos 'Ingreso_Total_Venta' y encontramos el que tiene el valor máximo.
print("26. Método de Pago con el mayor Ingreso Total de Venta:")
metodo_pago_top_ingreso = df_ventas.groupby('Metodo_Pago')['Ingreso_Total_Venta'].sum().idxmax()
ingreso_metodo_pago = df_ventas.groupby('Metodo_Pago')['Ingreso_Total_Venta'].sum().max()
print(f"El método de pago con el mayor ingreso total es: '{metodo_pago_top_ingreso}' con un ingreso de ${ingreso_metodo_pago:,.2f}")
print("-" * 50)

# --- 27. Aplicar Función Personalizada (Segmento de Edad): ---
# Define una función para clasificar la edad.
def clasificar_edad(edad):
    if edad < 30:
        return 'Joven'
    elif 30 <= edad <= 50:
        return 'Adulto'
    else:
        return 'Mayor'

print("27. Calculando la Ganancia Bruta promedio por Segmento de Edad:")
# Aplica la función a la columna 'Cliente_Edad' para crear 'Segmento_Edad'.
df_ventas['Segmento_Edad'] = df_ventas['Cliente_Edad'].apply(clasificar_edad)
# Agrupa por el nuevo segmento de edad y calcula la media de la ganancia bruta.
ganancia_por_segmento_edad = df_ventas.groupby('Segmento_Edad')['Ganancia_Bruta'].mean()
print(ganancia_por_segmento_edad)
print("-" * 50)

# --- 28. Tendencia de Ventas Mensuales: ---
# Extraemos el número del mes de la columna 'Fecha'.
print("28. Tendencia de Ingresos Totales de Venta Mensuales:")
df_ventas['Mes_Venta'] = df_ventas['Fecha'].dt.month # .dt.month extrae el mes de una columna datetime.
# Agrupamos por 'Mes_Venta' y sumamos el 'Ingreso_Total_Venta'.
ingreso_mensual = df_ventas.groupby('Mes_Venta')['Ingreso_Total_Venta'].sum()
print(ingreso_mensual)
print("-" * 50)

# --- 29. Diagrama de Dispersión (Edad vs. Ganancia): ---
print("29. Creando Diagrama de Dispersión: Cliente_Edad vs. Ganancia_Bruta...")
plt.figure(figsize=(10, 6)) # Define el tamaño del gráfico.
plt.scatter(df_ventas['Cliente_Edad'], df_ventas['Ganancia_Bruta'], alpha=0.6) # Crea el gráfico de dispersión.
plt.title('Relación entre Edad del Cliente y Ganancia Bruta por Venta') # Título del gráfico.
plt.xlabel('Edad del Cliente') # Etiqueta del eje X.
plt.ylabel('Ganancia Bruta ($)') # Etiqueta del eje Y.
plt.grid(True, linestyle='--', alpha=0.7) # Añade una cuadrícula para mejor lectura.
plt.tight_layout() # Ajusta el diseño para que no se superpongan los elementos.
plt.show() # Muestra el gráfico.
print("Análisis de la dispersión: Observa si hay una concentración de puntos en ciertas edades o ganancias, o si hay una tendencia (positiva, negativa, nula).")
print("-" * 50)

# --- 30. Gráfico de Barras (Ingreso por Categoría): ---
print("30. Creando Gráfico de Barras: Ingreso Total por Categoría de Producto...")
# Primero, agrupamos y sumamos el ingreso por categoría.
ingreso_por_categoria = df_ventas.groupby('Categoria')['Ingreso_Total_Venta'].sum()
# Luego, ordenamos las categorías de mayor a menor ingreso.
ingreso_por_categoria_ordenado = ingreso_por_categoria.sort_values(ascending=False)

plt.figure(figsize=(12, 7)) # Define el tamaño del gráfico.
# Crea el gráfico de barras. `ingreso_por_categoria_ordenado.index` son los nombres de las categorías.
# `ingreso_por_categoria_ordenado.values` son los ingresos correspondientes.
plt.bar(ingreso_por_categoria_ordenado.index, ingreso_por_categoria_ordenado.values, color='skyblue')
plt.title('Ingreso Total de Venta por Categoría de Producto') # Título del gráfico.
plt.xlabel('Categoría de Producto') # Etiqueta del eje X.
plt.ylabel('Ingreso Total de Venta ($)') # Etiqueta del eje Y.
plt.xticks(rotation=45, ha='right') # Rota las etiquetas del eje X para mejor legibilidad si son largas.
plt.grid(axis='y', linestyle='--', alpha=0.7) # Añade una cuadrícula solo en el eje Y.
plt.tight_layout() # Ajusta el diseño.
plt.show() # Muestra el gráfico.
print("Análisis del gráfico de barras: Puedes identificar rápidamente las categorías de productos que más contribuyen al ingreso total.")
print("-" * 50)
```
